# Inferência Estatística

A inferência estatística é um ramo da estatística que busca tirar conclusões sobre uma população com base em dados coletados de uma **amostra representativa** desta população.




AED.png


## Objetivos da Inferência Estatística

- **Estimar parâmetros** desconhecidos (características) de uma **população**, por meio de informações fornecidas pela **amostra**.

- **Testar hipóteses sobre os parâmetros populacionais**, por meio de evidências fornecidas pela **amostra**.

- Fazer **previsões** sobre valores futuros com base em dados históricos.

- Identificar **associação** entre variáveis de uma população.



## Métodos de Inferência Estatística

1. **Estimação Estatística:** Busca responder perguntas como:

- Qual a estimativa pontual de uma quantidade populacional, com base na amostra?

- Como obter um intervalo de confiança para determinada característica da população, com informações provenientes da amostra?


2. **Teste de Hipóteses:** Busca responder perguntas como:

- Como avaliar se um parâmetro da população pode ser considerado um determinado valor?

- Como avaliar se um tratamento tem efeitos diferentes nas pessoas?

- Como avaliar a associação entre variáveis a partir de uma amostra?




# Abordagens de Inferência Estatística

## Inferência Frequentista:

- Baseia-se na **frequência relativa** dos eventos em amostras grandes.

- Usa métodos de estimação como mínimos quadrados, método de momentos, máxima verossimilhança e para estimar parâmetros populacionais.

- Não incorpora **conhecimento prévio** sobre os parâmetros.

- Foca na probabilidade dos dados, conhecidos os parâmetros.

- Utiliza **intervalos de confiança e testes de hipóteses** para fazer conclusões sobre os parâmetros.


## Inferência Bayesiana:

- Incorpora **conhecimento prévio** sobre os parâmetros em forma de distribuição de probabilidade (priori).

- **Atualiza essa distribuição** com base nos dados observados usando o teorema de Bayes, resultando em uma distribuição **a posterior** para os parâmetros, que combina os dados e o conhecimento a priori.

- Fornece uma distribuição completa para os parâmetros, refletindo a incerteza associada.

- Foca na **probabilidade dos parâmetros**, conhecidos os dados.

- Permite a atualização contínua das crenças à medida que mais dados são coletados (atualização bayesiana).

## Teste  Z:

O teste Z é usado quando a **variância populacional é conhecida e a amostra é grande**. As suposições para o teste Z incluem:

- Distribuição normal e/ou tamanho da amostra grande o suficiente, a distribuição amostral das médias deve se aproximar da normal.

- O desvio padrão populacional ($\sigma$) deve ser conhecido.


A estatística de teste Z é calculada da seguinte forma:

$Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}$

Onde:
- $\bar{X}$ é a média amostral,
- $\mu$ é a média populacional,
- $\sigma$ é o desvio padrão populacional, e
- $n$ é o tamanho da amostra.


## Teste t de Student:

O teste t de Student é usado quando **a variância populacional é desconhecida** e/ou a **amostra é pequena**. As suposições para o teste t de Student incluem:

- As observações devem ser independentes.

- A população deve ser normalmente distribuída ou, quando o tamanho da amostra é pequeno, a distribuição amostral das médias deve se aproximar da normal.

- O desvio padrão populacional é desconhecido e estimado pelo desvio padrão amostral ($s$).


A estatística de teste t é calculada da seguinte forma:

$t = \displaystyle\frac{\bar{X} - \mu}{\frac{s}{\sqrt{n}}}$

Onde:
- $\bar{X}$ é a média amostral,
- $\mu$ é a média populacional,
- $s$ é o desvio padrão amostral, e
- $n$ é o tamanho da amostra.


## Análise de Variância (ANOVA):

A ANOVA é usada para comparar as médias de três ou mais grupos independentes. As suposições para a ANOVA incluem:

- As observações dentro de cada grupo são independentes e identicamente distribuídas.

- Cada grupo deve ter uma distribuição normal.

- Homogeneidade das variâncias: as variâncias nos grupos são aproximadamente iguais.


A estatística de teste F é calculada como a razão das variâncias entre os grupos e dentro dos grupos. A equação para a estatística de teste F é dada por:

$F = \frac{MS_{\text{entre}}}{MS_{\text{dentro}}}$

Onde:
- $MS_{\text{entre}}$ é a média dos quadrados entre os grupos,

- $MS_{\text{dentro}}$ é a média dos quadrados dentro dos grupos.


## Teste a média de uma população com a variância conhecida

$X_i \sim N(\mu, \sigma^2)$ com $\sigma^2$ conhecido

$X_1, \ldots, X_n$ amostra aleatória

Considere avaliar

$H_0: \mu = \mu_0$ contra

$H_1: \mu < \mu_0$

A estatística de teste é $\bar{X}$

A região crítica é

$RC = \{\bar{x} \in \mathbb{R}; \bar{x}< x_c\}$

Exemplo

$H_0: \mu =18$ contra

$H_1: \mu<18$

$RC = \{\bar{x} \in \mathbb{R}; \bar{x}< x_c\}$

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Exemplo

mu0 = 18
sigma = 6
n=30
xc = norm.ppf(0.05)*(sigma/np.sqrt(n)) + mu0

print('RC = {xbarra <', np.round(xc,2), '}')
```

```python
# xbarra observado e decisão do teste

xbarra = 16

if xbarra < xc: print ('Rejeito H0')
```

```python
# Nível descritivo do teste

z_obs = (xbarra - mu0)/(sigma/np.sqrt(n))

print('valor-p', np.round(norm.cdf(z_obs),3))
```

## Teste de normalidade dos dados: Shapiro Wilk

#### Atenção! Para usar testes paramétricos, verificar se é necessário testar a normalidade dos dados antes!

$X_1, \ldots, X_n$ amostra aleatória

$H_0: X_i \sim Normal$ contra

$H_1:  X_i $ tem outra distribuição


```python
# Fonte: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html

from scipy import stats
rng = np.random.default_rng()
x = stats.norm.rvs(loc=5, scale=3, size=100, random_state=rng)
shapiro_test = stats.shapiro(x)
shapiro_test
```

```python
shapiro_test[0]
```

```python
shapiro_test.pvalue
```

```python
#decisão do teste

if shapiro_test.pvalue < 0.05: print ('Rejeito H0')
else: print ('Não rejeito H0')
```

## Teste a média de uma população com a variância desconhecida

$X_i \sim N(\mu, \sigma^2)$ com $\sigma^2$ desconhecido

$X_1, \ldots, X_n$ amostra aleatória




Considere avaliar

$H_0: \mu =\mu_0$ contra

$H_1: \mu>\mu_0$



A estatística de teste é $\bar{T} = \frac{\bar{X}-\mu}{S/\sqrt{n}}$

A região crítica é



$RC = \{T \in \mathbb{R}; T > t_c\}$

Exemplo:

$H_0: \mu =30$ contra

$H_1: \mu>30$


```python
from scipy.stats import t
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Dados do exercício

mu0 = 30
s = 3
n=25

tc = mu0-t.ppf(0.05, n-1)*(s/np.sqrt(n))

print('RC = {t >', np.round(tc,2), '}')
```

```python
# t observado e decisão do teste

tobs = 31.5

if tobs > tc: print ('Rejeito H0')
else: print('Não rejeito H0')
```

```python
# Nível descritivo do teste

print('valor-p', np.round(t.cdf(z_obs, n-1),3))
```

## Teste para variância de uma população

Considere avaliar

$H_0: \sigma^2 = \sigma^2_0$ contra

$H_1: \sigma^2 \neq \sigma^2_0$.

Usamos a estatística baseada na variância amostral observada $s^2$

$v = \displaystyle\frac{(n-1)s^2}{\sigma^2_0}$

$RC = \{v \in \mathbb{R}^+; v < v_{c1} \mbox{  ou  }  v > v_{c2}\}$

```python
from scipy.stats import chi2

# Dados do exercício


s2 = 1.5
sigma2_0 = 1
n = 20

vobs = (n-1) * s2 / sigma2_0

vc1 = chi2.ppf(0.025, n-1)
vc2 = chi2.ppf(0.975, n-1)

print('RC = {v <', np.round(vc1,2), 'ou v >', np.round(vc2,2), '}')

if ((vobs > vc2) or (vobs < vc1)) : print ('Rejeito H0')
else: print('Não rejeito H0.')
```

## Teste F para igualdade de variâncias de duas populações

Considere avaliar

$H_0: \sigma^2_1 = \sigma^2_2$ contra

$H_1: \sigma^2_1 \neq \sigma^2_2$.

Usamos a estatística baseada nas variâncias amostrais observadas $s^2_1$ e $s^2_2$.

$F = \displaystyle\frac{s^2_1}{s^2_2}$

$RC = \{F \in \mathbb{R}^+; F < f_{c1} \mbox{  ou  }  F > f_{c2}\}$

```python
from scipy.stats import f

# Dados do exercício

s2_1 = 1500
s2_2 = 2500

n1 = 26
n2 = 21

Fobs = s2_1 / s2_2

fc1 = f.ppf(0.025, n1-1, n2-1)
fc2 = f.ppf(0.975, n1-1, n2-1)

print('RC = {F <', np.round(fc1,2), 'ou F >', np.round(fc2,2), '}')
print('Fobs=',np.round(Fobs,2))


if ((Fobs > fc2) or (Fobs < fc1)) : print ('Rejeito H0')
else: print('Não rejeito H0.')
```

## Análise de variância ANOVA

Considere o modelo de ANOVA com um fator

$$Y_{ij} = \mu_{i} + \epsilon_{ij}, i=1,\ldots,k; j=1,\ldots,m.$$

em que $m$ é o número de observações por grupo e $k$ é o número de grupos.

Queremos avaliar se

$H_0: \mu_1=\mu_2=\ldots \mu_k$ contra

$H_1:$ pelo menos uma média diferente das outras.

Suposições:

- $Y_{ij}$ são v.a.'s independentes
- Todas as k populações tem variâncias iguais $\sigma^2$ (homocedasticidade)
- $Y_{ij}\sim N(\mu, \sigma^2)$ para $i=1,\ldots,k$ e $j=1,\ldots,m$.


Exemplo: Deseja-se avaliar se a média de uma métrica de avaliação de desempenho é a mesma em 4 tipos de máquinas diferentes. Para isso, foram realizados testes em 7 máquinas de cada tipo e a métrica foi calculada.

Podemos usar a função scipy.stats.f_oneway.

```python
import pandas as pd

dados = pd.read_csv('http://raw.githubusercontent.com/cibelerusso/IntroducaoaInferenciaEstatistica/main/Dados/desempenho.csv', sep=';')
dados
```

```python
dados.describe()
```

```python
import matplotlib.pyplot as plt

dados.plot.box()
```

```python
# Verificando a normalidade dos dados

from scipy import stats

shapiro_test, p1 = stats.shapiro(dados['Tipo 1'])
shapiro_test, p2 = stats.shapiro(dados['Tipo 2'])
shapiro_test, p3 = stats.shapiro(dados['Tipo 3'])
shapiro_test, p4 = stats.shapiro(dados['Tipo 4'])

print(p1, p2, p3, p4)

# Vamos supor, para a aplicação do teste, que é razoável supor a normalidade e que os dados tem a mesma variância -
```

```python
from scipy.stats import f_oneway

F, p = f_oneway(dados['Tipo 1'], dados['Tipo 2'], dados['Tipo 3'], dados['Tipo 4'])

print(F, p)
```

```python
# Tabela ANOVA - Com base de dados expandida

from statsmodels.formula.api import ols
import statsmodels.api as sm

# Fonte: https://www.reneshbedre.com/blog/anova.html
# reshape the d dataframe suitable for statsmodels package

dados_melt = pd.melt(dados.reset_index(), id_vars=['index'], value_vars=['Tipo 1', 'Tipo 2', 'Tipo 3', 'Tipo 4'])

# replace column names
dados_melt.columns = ['index', 'Tipo', 'Métrica']
dados_melt
```

## E a modelagem estatística?

- A modelagem estatística busca compreender a relação entre variáveis.

- Os modelos estatísticos se diferenciam dos modelos de machine learning, por exemplo, pois permitem fazer **inferência estatística** sobre os coeficientes (parâmetros) do modelo.

- Podem não alcançar poder preditivo tão alto quando comparado com alguns modelos mais complexos como de redes neurais artificiais, mas ganham em **interpretabilidade** e **poder de extrapolação**, o que é muito importante em resultados de modelagem.
